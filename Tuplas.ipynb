#Tuplas
Características:
- Seus elementos possuem índices
- Podem ser heterrogêneas (possuir tipos diferentes de dados)
- É imutável
- Tupla = (elemento1, elemento2, elemento3...)

```python
pessoas = ('Giovanna', 17, 'Isabella', 21, 'Gabriel', 18)
```

```python
type(pessoas)
```

```python
pessoas[0]
```

```python
tupla = ('Giovanna', 17, ['Demon Slayer', 'One Piece', 'Dragon Ball-z'])
```

```python
type(tupla[2])
```

```python
tupla[2][1] = 'Jujutsu Kaisen'
```
