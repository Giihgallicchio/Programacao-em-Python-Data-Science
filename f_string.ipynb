# f-strings
- Também chamadas de strings literais formatadas (formatted string literais).
- f-strings são strings com a letra F no início e chaves {} para realizar a interpolação de variáveis ou expressões.

```python
idade = 17
```

```python
print('Meu nome é Giovanna e eu tenho', idade, 'anos.')
```

```python
# f-string
idade = 17
print(f'Meu nome é Giovanna e eu tenho {idade} anos.')
```

#Slicing (fatiamento)
- A técnica de fatiamento nos permite acessar partes de uma string utilizando a notação de uma índice

```python
linha = '10010GIOVANNAALVESGALICHIORUADOUTORABERLADODACUNHALOBO2111965512762'
```

```python
print(f'Cód. Cliente: {linha[0:5]} - Nome Cliente: {linha[5:26]} - Endereço do Cliente: {linha[26:54]} - Nº da Casa: {linha[54:56]} - Nº de Telefone: {linha[56:70]}')
```
