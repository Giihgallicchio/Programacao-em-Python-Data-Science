# Sets (Conjuntos)
- São mutáveis
- Elementos não possui índices
- Não contém elementos repetidos
- Permitem operações em conjuntos

```python
conjunto1 = {1,2,3,4,5}
```

```python
conjunto2 = {5,6,7,8,9}
```

```python
#O union ele junta os dois conjuntos, ele também nao duplica se um número estiver repetido
conjunto1.union(conjunto2)
```

```python
#Ele mostra o número repetido
conjunto1.intersection(conjunto2)
```

```python
#Ele mostra a diferença do conjunto1 para o conjunto2
conjunto1.difference(conjunto2)
```

```python
#Ele mostra a diferença do conjunto2 para o conjunto1, o numero 5 nao aparece por conta que ele esta nos dois conjuntos
conjunto2.difference(conjunto1)
```

```python
conjunto1.add(10)
```
