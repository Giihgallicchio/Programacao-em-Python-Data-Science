# Estrutura condicional **IF**
Executar ou não um comando ou um bloco de comandos de acordo com o resultado de um teste condicional (Teste Lógico)

- if significa "se"
- Testes lógicos podem utilizar os operadores lógicos (and, or ou not)
- Estrutura condicional pode ser simples ou composta (else - "senão")

```python
# Estrutura condicional simples
faturamento = 150000
tempo_empresa = 1
if faturamento >= 100000 and tempo_empresa >=2:
  print(faturamento * 0.03)
```

```python
# Estrutura condicional composta
faturamento = 150000
tempo_empresa = 1
if faturamento >= 100000 and tempo_empresa >=2:
  print(faturamento * 0.03)
else:
  print(faturamento * 0.01)
```

```python
# Estrutura condicional composta
faturamento = 98000
tempo_empresa = 0
if faturamento >= 100000 and tempo_empresa >=2:
  print(faturamento * 0.03)

elif faturamento >= 90000:
  print(faturamento * 0.02)

else:
  print(faturamento * 0.01)
```
