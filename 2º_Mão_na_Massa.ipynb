```python
df = df_ordenado = df.sort_values(by='Lucro',ascending=True)
df_ordenado
```

```python
df_ordenado.groupby(['Segmento', 'País', 'Ano'])[['Lucro']].min().reset_index().sort_values(by='Lucro', ascending=True) #Lucro Negativo
```

```python
df_ordenado.groupby(['Segmento', 'País', 'Ano'])[['Lucro']].min().reset_index().sort_values(by='Lucro', ascending=False) #Lucro Positivo
```
