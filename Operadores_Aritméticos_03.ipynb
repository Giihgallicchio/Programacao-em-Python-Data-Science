# Operadores Aritméticos
No Python também temos operadores para realizarmos cálculos, desde os mais simples até os mais complexos.

```python
# Adição
print(10 + 20 + 30)
```

```python
# Subtração
print(123 - 87)
```

```python
# Multiplicação
print(10 * 8)
```

```python
#Divisão
print(100 / 10)
```

```python
# Raiz quadrada
print(9 ** 0.5)
```

```python
# Potência
print(2 ** 3)
```

```python
# Divisão com resultado inteiro
print(129 / 23)
print(129 // 23) #Duas barras ele arredonda o resultado
```

```python
# Módulo (Resto da Divisão)
print(129 % 23)
```
