# Preparação de Dados (Data Prep)
### Tratando Dados Duplicados

```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
        'Aluno': ['João', 'Maria', 'Helena', 'Joel', 'Gabriel'],
        'Idade': [30,25,18,29,45]
    }
)
df
```

```python
#Mostra na onde esta duplicado
df.duplicated()
```

```python
#Tras a quantidade de duplicidade
df.duplicated().sum()
#tambem pode mostrar uma coluna em especifico a quatindade de duplicidade
df['Aluno'].duplicated().sum()
```

```python
#Duplicando as informações
df_dup = pd.concat( (df, df) )
#ordenando o nome do aluno
df_dup.sort_values(by='Aluno')
```

```python
df_dup.duplicated().sum()
```

```python
df_dup['Aluno'].duplicated().sum()
```

```python
#exclui a duplicidade
df_unicos = df_dup.drop_duplicates()
df_unicos
```
