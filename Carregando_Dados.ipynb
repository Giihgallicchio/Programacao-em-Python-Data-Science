```python
# Arquivos de Banco de Dados -SQLite3
conn = sqlite3.connect('DataSets/vendas.db')
df = pd.read_sql_query('Select descricao_produto, count(*) as Qtde \
                       From pedidos Group By descricao_produto \
                       Order By count(*) desc ', conn)
df.head()
```
